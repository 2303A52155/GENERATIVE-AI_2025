 RESUME AND COVER LETTER GENERATOR

```python
import pandas as pd
from datetime import date

# Load the datasets
resume_df = pd.read_csv("/content/UpdatedResumeDataSet.csv")
cover_letter_df = pd.read_csv("/content/cover_letter_dataset.csv")

# Function to generate a formatted resume
def generate_resume(name, email, phone, education, skills, job_title, certifications, achievements,
                    extracurricular, strengths, languages, projects, internships):
    resume = f"""
ðŸ“„ RESUME

--------------------------------
1. PERSONAL DETAILS
--------------------------------
Name: {name}
Email: {email}
Phone: {phone}

--------------------------------
2. CAREER OBJECTIVE
--------------------------------
Seeking a challenging role as a {job_title} to apply my skills and contribute to organizational success.

--------------------------------
3. EDUCATIONAL QUALIFICATIONS
--------------------------------
{education}

--------------------------------
4. TECHNICAL SKILLS
--------------------------------
{skills}

--------------------------------
5. PROJECTS
--------------------------------
{projects}

--------------------------------
6. INTERNSHIPS
--------------------------------
{internships}

--------------------------------
7. CERTIFICATIONS
--------------------------------
{certifications}

--------------------------------
8. ACHIEVEMENTS
--------------------------------
{achievements}

--------------------------------
9. EXTRA-CURRICULAR ACTIVITIES
--------------------------------
{extracurricular}

--------------------------------
10. PERSONAL STRENGTHS
--------------------------------
{strengths}

--------------------------------
11. LANGUAGES KNOWN
--------------------------------
{languages}

--------------------------------
12. DECLARATION
--------------------------------
I hereby declare that all the above details are true to the best of my knowledge.
    """
    return resume

# Function to generate a formatted cover letter
def generate_cover_letter(name, email, phone, job_title, company_name, address, education, skills, achievements):
    today = date.today().strftime("%B %d, %Y")
    letter = f"""
ðŸ“„ COVER LETTER

{name}
Email: {email}
Phone: {phone}
Date: {today}

Hiring Manager
{company_name}
{address}

Subject: Application for the Position of {job_title}

Dear Hiring Manager,

I am writing to express my interest in the position of {job_title} at {company_name}. With my educational background in {education} and a strong skill set including {skills}, I am confident in my ability to contribute meaningfully to your team.

Throughout my academic and personal development, I have focused on honing my technical capabilities, particularly in areas relevant to the {job_title} role. I have consistently demonstrated a passion for learning and applying new concepts and have worked on projects and tasks that reflect my dedication and adaptability.

My achievements include {achievements}, which highlight my commitment to excellence and goal-oriented approach. I am eager to bring this same energy and determination to {company_name} and grow professionally while contributing to the success of the organization.

I would welcome the opportunity to further discuss how I can be an asset to your company. Please find my resume enclosed for additional details.

Thank you for your time and consideration.

Sincerely,
{name}
    """
    return letter

# Get user input dynamically
name = input("Enter your name: ")
email = input("Enter your email: ")
phone = input("Enter your phone number: ")
education = input("Enter your educational background: ")
skills = input("Enter your key technical skills (comma-separated): ")
job_title = input("Enter the job title you are applying for: ")
certifications = input("Enter your certifications: ")
achievements = input("Enter your achievements: ")
extracurricular = input("Enter your extra-curricular activities: ")
strengths = input("Enter your personal strengths: ")
languages = input("Enter languages known: ")
projects = input("Enter your projects (comma-separated): ")
internships = input("Enter your internships (comma-separated): ")
company_name = input("Enter the company name: ")
company_address = input("Enter the company address: ")

# Generate resume and cover letter
resume_output = generate_resume(name, email, phone, education, skills, job_title, certifications,
                                achievements, extracurricular, strengths, languages, projects, internships)
cover_letter_output = generate_cover_letter(name, email, phone, job_title, company_name,
                                             company_address, education, skills, achievements)

# Print results
print("\n====== GENERATED RESUME ======\n")
print(resume_output)

print("\n====== GENERATED COVER LETTER ======\n")
print(cover_letter_output)

# Show example resumes from the dataset matching the job title category
print("\n====== EXAMPLE RESUMES FROM DATASET ======\n")
category_matches = resume_df[resume_df['Category'].str.contains(job_title, case=False, na=False)]
examples = category_matches['Resume'].sample(min(3, len(category_matches)))
for i, resume in enumerate(examples, 1):
    print(f"\n--- Example {i} ---\n{resume[:1000]}\n...")
```
