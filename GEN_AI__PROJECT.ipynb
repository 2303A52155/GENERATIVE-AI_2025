```python
import pandas as pd
import random
from datetime import date

# Load the datasets
resume_df = pd.read_csv("/content/UpdatedResumeDataSet.csv")
cover_letter_df = pd.read_csv("/content/cover_letter_dataset.csv")

# Function to generate a formatted resume
def generate_resume(name, email, phone, education, skills, job_title, certifications, achievements,
                    extracurricular, strengths, languages):
    resume = f"""
ðŸ“„ RESUME

--------------------------------
1. PERSONAL DETAILS
--------------------------------
Name: {name}
Email: {email}
Phone: {phone}

--------------------------------
2. CAREER OBJECTIVE
--------------------------------
Seeking a challenging role as a {job_title} to apply my skills and contribute to organizational success.

--------------------------------
3. EDUCATIONAL QUALIFICATIONS
--------------------------------
{education}

--------------------------------
4. TECHNICAL SKILLS
--------------------------------
{skills}

--------------------------------
5. PROJECTS
--------------------------------
- Freelancing Opportunities Platform
- Library Management System

--------------------------------
6. INTERNSHIPS
--------------------------------
- Web Development Intern

--------------------------------
7. CERTIFICATIONS
--------------------------------
{certifications}

--------------------------------
8. ACHIEVEMENTS
--------------------------------
{achievements}

--------------------------------
9. EXTRA-CURRICULAR ACTIVITIES
--------------------------------
{extracurricular}

--------------------------------
10. PERSONAL STRENGTHS
--------------------------------
{strengths}

--------------------------------
11. LANGUAGES KNOWN
--------------------------------
{languages}

--------------------------------
12. DECLARATION
--------------------------------
I hereby declare that all the above details are true to the best of my knowledge.
    """
    return resume

# Function to generate a formatted cover letter
def generate_cover_letter(name, email, phone, job_title, company_name, address):
    today = date.today().strftime("%B %d, %Y")
    letter = f"""
ðŸ“„ COVER LETTER

{name}
12-3/45, Ameerpet,
Hyderabad, Telangana â€“ 500016
Email: {email}
Phone: {phone}
Date: {today}

Hiring Manager
{company_name}
{address}

Subject: Application for the Position of {job_title}

Dear Hiring Manager,

I am writing to express my keen interest in the position of {job_title} at {company_name}, as advertised. With a solid academic background in Computer Science and Engineering, coupled with hands-on experience in software development and problem-solving, I am confident in my ability to contribute effectively to your dynamic team.

I recently completed my B.Tech in Computer Science from XYZ Institute of Technology, Hyderabad, where I maintained a CGPA of 8.5. During my academic journey, I worked on several projects that sharpened my technical and collaborative skills, including a freelancing opportunities platform, which was selected for the Smart India Hackathon 2025 Finals. This project enhanced my abilities in backend development, database management, and UI integration.

In addition to academic projects, I also completed a web development internship at CodeAlpha Technologies, where I developed user-friendly web pages and worked on form validation and API integration. I am proficient in programming languages like Java, C, and Python and have experience using tools such as Git, VS Code, and MySQL.

What excites me about {company_name} is your commitment to innovation and your emphasis on delivering high-quality solutions to real-world problems. I am highly motivated to join an organization where I can apply my skills in a challenging environment and grow continuously by learning from industry leaders.

Enclosed is my resume, which provides further details about my academic and technical background. I look forward to the opportunity of discussing how I can contribute to your esteemed organization in person. Thank you for considering my application.

Warm regards,
{name}
    """
    return letter

# Get user input dynamically
name = input("Enter your name: ")
email = input("Enter your email: ")
phone = input("Enter your phone number: ")
education = input("Enter your educational background: ")
skills = input("Enter your key technical skills (comma-separated): ")
job_title = input("Enter the job title you are applying for: ")
certifications = input("Enter your certifications: ")
achievements = input("Enter your achievements: ")
extracurricular = input("Enter your extra-curricular activities: ")
strengths = input("Enter your personal strengths: ")
languages = input("Enter languages known: ")
company_name = input("Enter the company name: ")
company_address = input("Enter the company address: ")

# Generate resume and cover letter
resume_output = generate_resume(name, email, phone, education, skills, job_title, certifications,
                                achievements, extracurricular, strengths, languages)
cover_letter_output = generate_cover_letter(name, email, phone, job_title, company_name, company_address)

# Print results
print("\n====== GENERATED RESUME ======\n")
print(resume_output)

print("\n====== GENERATED COVER LETTER ======\n")
print(cover_letter_output)

# Show example resumes from the dataset matching the job title category
print("\n====== EXAMPLE RESUMES FROM DATASET ======\n")
category_matches = resume_df[resume_df['Category'].str.contains(job_title, case=False, na=False)]
examples = category_matches['Resume'].sample(min(3, len(category_matches)))
for i, resume in enumerate(examples, 1):
    print(f"\n--- Example {i} ---\n{resume[:1000]}\n...")
```
